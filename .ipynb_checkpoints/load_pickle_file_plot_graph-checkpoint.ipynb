```python
import pickle
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Detective Game

## Neuro Agent Strategy

```python
with open('./track_scores_detective_pop50gen500_170818.pkl', 'rb') as i:
    x = pickle.load(i)
```

```python
x[:2]
# Columns are counters and accumulated rewards
```

```python
df = pd.DataFrame(x, columns=['State','Population no.','Generation No.','Action','Reward to Game Agent','Game Reward','Accumulated Reward'])
```

```python
df[:5]
```

# Seaborn

```python
#sns.set(style="darkgrid")

# Plot the responses for different events and regions
sns_plot = sns.lineplot(x="Generation No.", y="Accumulated Reward",
             hue="Population no.",
             data=df)
fig = sns_plot.get_figure()
fig.savefig('Detective - Errorband Plot by Population.png',bbox_inches='tight')
#sns_plot.savefig("detective_plot_seaborn_neuro.png",bbox_inches='tight')
```

```python
# Group dataset by name of place and year, calculate the average amount of snow for each group
df["accumulated_rewards_average"] = df.groupby(['Generation No.'])['Accumulated Reward'].transform("mean")
#df["accumulated_rewards_average"] = df.groupby(['Generation No.'])['Game Reward'].transform("mean")
```

```python
plt.ylim(0,80)
plt.plot(df['Generation No.'], df['accumulated_rewards_average'])
plt.ylabel('Accumulated Reward')
plt.xlabel('Generations')
plt.savefig('Evolution - Detective - Mean Accumulated Reward by Generations.png')

plt.show()
```

# Load words seen in the game pickle

```python
import word2vec, sys, os, math
```

```python
word2vec_bin_loc = 'scholar/untagged_wikipedia_for_word2vec.bin'
```

```python
def load_word2vec(model_filename):
    model = word2vec.load(model_filename)
    return model
```

```python
word_vectors = load_word2vec(word2vec_bin_loc)
```

```python
import pickle as pkl

with open('./words_seen_detective_pop10gen100_trial_160818.pkl', 'rb') as i:
    x = pkl.load(i)
```

```python
labels = []
tokens = []

for y in x:
    label = x[y]
    tokens.append(word_vectors.vectors[label])
    labels.append(word_vectors.vocab[label])
```
