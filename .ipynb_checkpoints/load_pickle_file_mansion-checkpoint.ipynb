## Neuroevolution Strategy

```python
with open('./runs/track_scores_cavetrip_1150718.pkl', 'rb') as i:
    y = pickle.load(i)
```

```python
print(y)
# Counter, Generation No, Asked NO, Accumulated Rewards
```

```python
df1 = pd.DataFrame(y, columns=['Counters','Generation No', 'Asked NO','Accumulated Reward'])
```

```python
# Group dataset by name of place and year, calculate the average amount of snow for each group
df1["accumulated_rewards_average"] = df1.groupby(['Counters'])['Accumulated Reward'].transform("mean")
```

```python
plt.ylim(0,80)
plt.plot(df1['Counters'], df1['accumulated_rewards_average'])
plt.ylabel('Scores')
plt.xlabel('Counters')
#plt.savefig('Image2.png')

plt.show()
```
